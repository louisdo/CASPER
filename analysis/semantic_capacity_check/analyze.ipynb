```python
import json
from collections import Counter
from tqdm import tqdm
from transformers import AutoTokenizer
```

```python
tokenizer = AutoTokenizer.from_pretrained("/scratch/lamdo/phrase_splade_checkpoints/phrase_splade_39/debug/checkpoint/model")
```

```python
vocab = tokenizer.vocab
reverse_voc = {v:k for k,v in vocab.items()}
```

```python
representations = []
with open("test_gitig_.json") as f:
    for line in tqdm(f):
        jline = json.loads(line)
        vector = {int(k): int(v) for k,v in jline.get("vector", {}).items()}

        title = jline.get("title")
        abstract = jline.get("text")

        doc = f"{title}. {abstract}"
        tokens = tokenizer(doc)["input_ids"]

        representations.append(Counter({k:v for k,v in vector.items() if k not in tokens}))
```

```python
representations[0]
```

```python
top_key_counter = Counter()
for line in tqdm(representations):
    top_key_counter.update([item[0] for item in line.items() if item[1]>=50])
```

```python
top = 80000
sum([1 if int(item[0]) >= 30522 else 0 for item in top_key_counter.most_common(top)]), len(top_key_counter.most_common(top))
```
