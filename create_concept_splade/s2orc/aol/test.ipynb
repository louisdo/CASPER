```python
import pandas as pd
import json, os
from transformers import AutoTokenizer
```

```python
tokenizer = AutoTokenizer.from_pretrained("lamdo/distilbert-base-uncased-phrase-60kaddedphrasesfroms2orc")
vocab = tokenizer.vocab
```

```python
df = pd.read_csv("/home/lamdo/splade/data/aol_keyphrase.csv")
df
```

```python
phrases = []
for line in df.to_dict(orient='records'):
    to_append = {"name": line["name"], "level": line["level"]}
    phrases.append(to_append)
```

```python
phrase_occurrence_folder = "/scratch/lamdo/s2orc_phrase_vocab"
files = os.listdir(phrase_occurrence_folder)
files = [os.path.join(phrase_occurrence_folder, file) for file in files if file.endswith(".json")]

for file in files:
    data = {}
    with open(file) as f:
        temp = json.load(f)

        for k,v in temp.items():
            if k not in data: data[k] = []

            data[k].extend(v)
```

```python
context_count = {}
for i, phrase in enumerate(phrases):

    count = len(data.get(phrase["name"], []))
    phrases[i]["count"] = count
```

```python
processed_phrases = [item for item in phrases if item["count"]>= 10]
```

```python
with open("test_gitig_.jsonl", "w") as f:
    for line in processed_phrases:
        json.dump(line, f)
        f.write("\n")
```
